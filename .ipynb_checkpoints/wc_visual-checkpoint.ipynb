```python
from data_cleaner import stopwords_corpus
from fetch_speech_to_df import spoke_persons
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sentiment_analysis import create_barplot

wc = WordCloud()

for i in range(4):
    wc.generate(stopwords_corpus.speeches[i])
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.title(spoke_persons[i])
    plt.show()

create_barplot()
```
